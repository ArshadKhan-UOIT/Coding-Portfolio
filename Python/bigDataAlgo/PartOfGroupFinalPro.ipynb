# **Converting Categorical attributes into numerical attributes**

```python
from sklearn import preprocessing
```

```python
lb = preprocessing.LabelEncoder()
```

```python
data_3['B_EMAIL'] = lb.fit_transform(data_3['B_EMAIL'])
data_3['CLASS'] = lb.fit_transform(data_3['CLASS'])
```

```python
data_3['FLAG_LRIDENTISCH'] = lb.fit_transform(data_3['FLAG_LRIDENTISCH'])
data_3['FLAG_NEWSLETTER'] = lb.fit_transform(data_3['FLAG_NEWSLETTER'])
data_3['CHK_LADR'] = lb.fit_transform(data_3['CHK_LADR'])
data_3['CHK_RADR'] = lb.fit_transform(data_3['CHK_RADR'])
data_3['CHK_KTO'] = lb.fit_transform(data_3['CHK_KTO'])
data_3['CHK_CARD'] = lb.fit_transform(data_3['CHK_CARD'])
data_3['CHK_COOKIE'] = lb.fit_transform(data_3['CHK_COOKIE'])
data_3['CHK_IP'] = lb.fit_transform(data_3['CHK_IP'])
data_3['FAIL_LPLZ'] = lb.fit_transform(data_3['FAIL_LPLZ'])
data_3['FAIL_LORT'] = lb.fit_transform(data_3['FAIL_LORT'])
data_3[ 'FAIL_LPLZORTMATCH'] = lb.fit_transform(data_3[ 'FAIL_LPLZORTMATCH'])
data_3['FAIL_RPLZ'] = lb.fit_transform(data_3['FAIL_RPLZ'])
data_3['FAIL_RORT'] = lb.fit_transform(data_3['FAIL_RORT'])
data_3['FAIL_RPLZORTMATCH'] = lb.fit_transform(data_3['FAIL_RPLZORTMATCH'])
data_3['NEUKUNDE'] = lb.fit_transform(data_3['NEUKUNDE'])
data_3['Z_METHODE'] = pd.get_dummies(data_3['Z_METHODE'])
data_3['WEEKDAY_ORDER'] = pd.get_dummies(data_3['WEEKDAY_ORDER'])
data_3['TIME_ORDER'] = pd.get_dummies(data_3['TIME_ORDER'])
```

```python
data_3.head()
```

```python
data_3['B_TELEFON'] = pd.get_dummies(data_3['B_TELEFON'])
data_3.head(5)
```
